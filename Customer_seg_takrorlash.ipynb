Customer segmentation (takroriy)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
url="https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/customer_segmentation.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Defaulted']=df['Defaulted'].fillna(0)
```

```python
df.isnull().sum()
```

As Adress column is not important, we drop this column

```python
df.drop(columns='Address', inplace=True)
```

```python
df.head(2)
```

```python
X=df.drop(columns='Customer Id').values
```

```python
X[:1]
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
X[:1]
```

```python
kmeans=KMeans(n_clusters=3, random_state=42, init='k-means++')
kmeans.fit(X)
```

```python
kmeans.labels_[:10]
```

```python
kmeans.cluster_centers_
```

```python
df.head(2)
```

```python
df['cluster']=kmeans.labels_
```

```python
df.sample(2)
```

```python
df.groupby('cluster').mean()
```
